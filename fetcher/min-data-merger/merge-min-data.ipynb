```python
dirs_stock = pathlib.Path('../../data/').glob('**/raw-minute')
```

```python
dirs = [x for x in dirs_stock if x.is_dir()]
```

```python
dirs
```

```python
def concat(df, df1):
  df = pd.concat([df, df1])
  return df.drop_duplicates().reset_index(drop=True)

for dir in dirs:
   p = pathlib.Path(dir).glob('data-*')
   files = [x for x in p if x.is_file()]
   df = None
   for file in files:
      df1 = pd.read_json(file)
      df = df1 if df is None else concat(df, df1)
   df = df[df['close'].notnull()]
   df = df.sort_values(by='timestamp')
   merged_path = dir/'../merged.csv'
   print(f'Processing {dir} Finished, saving to {merged_path}')
   print(f'Saving to {merged_path}')
   df.to_csv(merged_path, index=False)
```

```python
def concat_2_files(file1, file2):
  df1 = pd.read_json(file1)
  df2 = pd.read_json(file2)
  df = pd.concat([df1, df2])
  return df.drop_duplicates()
```
